### Prepare: smooth the daily noise

Daily counts are spiky. A **rolling average** replaces each day with the average of a window around it, revealing the trend. This one line is the kind of data preparation teachers do all the time, and `.idxmax()` returns the row label of the largest value, which the next cell hands to `az.loc[...]` to pull up that busiest day's full row. We use a 7-day window on purpose: flu reporting follows a weekly rhythm (fewer cases are logged on weekends), so a 7-day average cancels that day-of-week effect and leaves the real trend.

```python
# 9b
# Rolling average of the daily hospitalizations
WINDOW = 7   # Exercise 3c: change this to 14, rerun this cell, then rerun the plot cell
az[f"rolling_{WINDOW}"] = az["hospitalizations"].rolling(window=WINDOW, center=True).mean()
# center=True averages days before and after each day (a symmetric window)

# When was the peak?
peak = az.loc[az["hospitalizations"].idxmax()]
print("peak day:", peak["date"].date(), "with", int(peak["hospitalizations"]), "hospitalizations")

az[["date", "hospitalizations", f"rolling_{WINDOW}"]].head(10)
```

### 5. Data Visualization with Classroom-Style Data

Here is a gallery of common chart types you can reuse on your own classroom data. Each uses a small dataset that resembles what a teacher might collect: study time, quiz scores, lab group, and absences.

A good chart usually answers one question at a time:

- **Line plot:** How does something change over time?
- **Bar chart:** How do groups compare?
- **Histogram:** What is the distribution?
- **Scatter plot:** How are two numerical variables related?
- **Box plot:** How do distributions compare across groups?
- **Heat map/correlation table:** Which numerical variables move together?

```python
# 11b
rng = np.random.default_rng(seed=7)
n_students = 36

class_data = pd.DataFrame({
    "student_id": np.arange(1, n_students + 1), "study_minutes": rng.integers(20, 180, size=n_students), "absences": rng.integers(0, 6, size=n_students), "lab_group": rng.choice(["A", "B", "C"], size=n_students),
})

# Create a realistic score: study helps, absences hurt, and there is random variation.
class_data["quiz_score"] = (
    55
    + 0.18 * class_data["study_minutes"]
    - 2.5 * class_data["absences"]
    + rng.normal(0, 7, size=n_students)
).clip(0, 100).round(1)

class_data.head()
```

```python
# 11c
# Visualization example 1: distribution of quiz scores
plt.figure(figsize=(7, 4))
plt.hist(class_data["quiz_score"], bins=8, edgecolor="black")
plt.title("Distribution of Quiz Scores")
plt.xlabel("Quiz score")
plt.ylabel("Number of students")
plt.show()

# Visualization example 2: relationship between study time and score
plt.figure(figsize=(7, 4))
plt.scatter(class_data["study_minutes"], class_data["quiz_score"])
plt.title("Study Time vs. Quiz Score")
plt.xlabel("Study time in minutes")
plt.ylabel("Quiz score")
plt.grid(True, alpha=0.3)
plt.show()

# Visualization example 3: compare groups
mean_by_group = class_data.groupby("lab_group")["quiz_score"].mean()
plt.figure(figsize=(6, 4))
plt.bar(mean_by_group.index, mean_by_group.values)
plt.title("Average Quiz Score by Lab Group")
plt.xlabel("Lab group")
plt.ylabel("Average score")
plt.ylim(0, 100)
plt.show()
```

```python
# 11d
# Visualization example 4: box plot by group
fig, ax = plt.subplots(figsize=(6, 4))
class_data.boxplot(column="quiz_score", by="lab_group", ax=ax)
plt.title("Quiz Score Distribution by Lab Group")
plt.suptitle("")
plt.xlabel("Lab group")
plt.ylabel("Quiz score")
plt.show()

# Visualization example 5: simple correlation heat map
numeric_cols = ["study_minutes", "absences", "quiz_score"]
corr = class_data[numeric_cols].corr()

plt.figure(figsize=(5, 4))
plt.imshow(corr, vmin=-1, vmax=1)
plt.colorbar(label="Correlation")
plt.xticks(range(len(numeric_cols)), numeric_cols, rotation=45, ha="right")
plt.yticks(range(len(numeric_cols)), numeric_cols)
plt.title("Correlation Heat Map")
for i in range(len(numeric_cols)):
    for j in range(len(numeric_cols)):
        plt.text(j, i, f"{corr.iloc[i, j]:.2f}", ha="center", va="center")
plt.tight_layout()
plt.show()
```

### 6. Loops and Conditionals: the Building Blocks

Every graph so far visualized data that was handed to us, and every calculation used a NumPy one-liner that worked on a whole array at once. Underneath those one-liners, Python is really doing the same small step over and over. In this section we write those steps ourselves with two coding building blocks:

- A **loop** repeats the same action many times, once per item.
- A **conditional** (`if` / `elif` / `else`) chooses what to do based on a test.

These are the plain control-flow tools behind almost any program. We practice them on a few short lists and on the `class_data` grade book, one small idea at a time. Each idea starts with the tiniest possible example that you can read at a glance, then we try it once on the grade book. Tomorrow you will use a loop to step a model forward in time. `14a`

### The for-loop: do something to each item

A **for-loop** walks through a collection one item at a time and runs the same indented lines for each item. A very common pattern is to keep a running total in an **accumulator** variable: start it at zero, then add to it on every pass. Start with just four test scores so you can watch the running total grow one line at a time, then compare the result with the NumPy one-liner you already know.

```python
# 14b
# A for-loop walks through a list one item at a time.
scores = [7, 9, 6, 8]

total = 0                     # start an accumulator at zero
for score in scores:          # take one score at a time
    total = total + score     # add it to the running total
    print("added", score, "running total is now", total)

print("final total:", total, " average:", total / len(scores))

# The NumPy one-liner you already know gives the same total:
print("numpy check:", np.sum(scores))
```

`15a`

### The same loop on the grade book

The loop does not care how long the list is. Here is the exact same accumulator pattern, but instead of four typed-in numbers it walks every student's `study_minutes` in the grade book and adds them up. The NumPy one-liner `.sum()` does the same thing in one step.

```python
# 15b
total_minutes = 0                              # start the accumulator at zero
for minutes in class_data["study_minutes"]:   # one student's minutes at a time
    total_minutes = total_minutes + minutes   # add it to the running total

print("total study minutes across the class:", total_minutes)
print("class average minutes:", total_minutes / len(class_data))
print("numpy check:", class_data["study_minutes"].sum())
```

`16a`

### Building a result by hand, then checking with NumPy

Loops let us rebuild a NumPy one-liner step by step, which is a good way to see what the one-liner is really doing. Here we find the largest value by hand on five daily temperatures: assume the first value is the biggest, then update that guess whenever the loop meets a larger one. That update is our first example of a **condition** inside a loop, which we name and practice fully in the next section.

```python
# 16b
temps = [68, 71, 65, 74, 70]     # five daily temperatures

warmest = temps[0]               # assume the first value is the biggest so far
for value in temps:              # look at every value
    if value > warmest:          # is this one bigger than the best so far?
        warmest = value          # if so, remember it

print("warmest by hand:", warmest)
print("numpy check:   ", np.max(temps))
```

### Conditionals: test each value, then count

An **if / elif / else** block picks exactly one branch based on a test. Putting a conditional inside a loop lets us look at each value and **count** how many pass a test. Start small: five quiz scores, with a cutoff of 70 to pass. Each score is added to the "passed" or "failed" tally as the loop visits it.

```python
# 17b
quiz = [55, 82, 90, 68, 74]      # five quiz scores

passed = 0                       # one tally for each outcome
failed = 0
for score in quiz:               # look at one score at a time
    if score >= 70:              # did it clear the cutoff?
        passed = passed + 1
        print(score, "-> passed, passed count now", passed)
    else:
        failed = failed + 1
        print(score, "-> failed, failed count now", failed)

print("passed:", passed, " failed:", failed)
```

`18a`

### Classify into more than two groups on the grade book

With `elif` we can sort each value into more than two groups. Below we give a short helper a name so we can reuse it, then loop over every quiz score in the grade book and tally three bands: "low", "mid", and "high". A dictionary holds one running count per band. We finish with a check that the three bands add up to the total number of students.

```python
# 18b
def score_band(score):
    """Return 'low', 'mid', or 'high' for a quiz score using two cutoffs.

    Parameters
    ----------
    score : float
        A single quiz score.

    Returns
    -------
    str
        One of 'low', 'mid', or 'high'.
    """
    if score < 70:
        return "low"
    elif score < 85:
        return "mid"
    else:
        return "high"


buckets = {"low": 0, "mid": 0, "high": 0}   # a tally for each band
for score in class_data["quiz_score"]:
    band = score_band(score)
    buckets[band] = buckets[band] + 1        # add one to that band's count

print("students in each band:", buckets)
print("check, bands add up to total students:", sum(buckets.values()), "of", len(class_data))
```

`19a`

### The while-loop: repeat until a condition is met

A **for-loop** runs a fixed number of times, once per item. A **while-loop** instead keeps going as long as a condition stays true and stops the moment it turns false. That makes it the right tool when you do not know the count in advance. Imagine saving toward a goal of 100 dollars: you keep adding deposits week by week and stop as soon as the balance reaches the goal. We do not know up front how many weeks that takes, so a while-loop is the natural fit.

```python
# 19b
goal = 100
weekly_deposits = [15, 20, 10, 25, 30, 40]   # what you add each week

balance = 0
week = 0
while balance < goal:                         # keep going while still short of the goal
    balance = balance + weekly_deposits[week] # add this week's deposit
    print("after week", week + 1, "balance is", balance)
    week = week + 1                           # move to the next week

print("reached the goal of", goal, "dollars in", week, "weeks")
```
